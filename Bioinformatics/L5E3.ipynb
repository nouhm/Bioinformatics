```python
def LevDynamicProgramming(text1, text2):
  rows, cols = (len(text1)+1, len(text2)+1)
  arr = [[0 for i in range(cols)] for j in range(rows)]

  for i in range(1,rows):
    arr[i][0] = i
  for i in range(1,cols):
    arr[0][i] = i

  for i in range(1,rows):
    for j in range(1,cols):
      # print(i , j)
      if text1[i-1] == text2[j-1]:
        arr[i][j] = arr[i-1][j-1]
      else:
        arr[i][j] = min(arr[i][j-1], arr[i-1][j], arr[i-1][j-1]) + 1

  distance = arr[-1][-1]
  print(arr)
  print('\n'.join([''.join(['{:4}'.format(item) for item in row]) 
      for row in arr]))
  print(distance)
  LevTraceback(text1, text2, arr, rows, cols)
```

```python
def LevTraceback(inputText1, inputText2, arr, rows, cols):
  text1 = [" " for i in range(max(rows, cols)-1)]
  text2 = [" " for i in range(max(rows, cols)-1)]

  # rows = 8 and cols = 7
  # print(rows , cols)
  i = rows-1
  j = cols-1

  for cell in reversed(range(0,max(rows, cols)-1)):

    if min(arr[i][j-1], arr[i-1][j], arr[i-1][j-1]) == arr[i-1][j-1]:
      i = i-1
      j = j-1
      text1[cell] = inputText1[i]
      text2[cell] = inputText2[j]
      
    elif min(arr[i][j-1], arr[i-1][j], arr[i-1][j-1]) == arr[i-1][j]:
      i = i-1
      text1[cell] = inputText1[i]
      
    elif min(arr[i][j-1], arr[i-1][j], arr[i-1][j-1]) == arr[i][j-1]:
      j = j-1
      text2[cell] = inputText2[i]
      
      
  print(text1)
  print(text2)
```

```python
# kitten
# sittingg

def LevRecursive(text1, text2):

  if len(text1) == 0:
    return len(text2)
  if len(text2) == 0:
    return len(text1)

  if text1[-1] == text2[-1]:
    return LevRecursive(text1[:-1], text2[:-1])

  return 1 + min(LevRecursive(text1[:-1], text2[:-1]), LevRecursive(text1[:-1], text2) , LevRecursive(text1, text2[:-1]))
```

```python
sequence = input("DNA nucleotide sequence: ") 
```

```python
def Count(seq):
  NUC = ["A", "T", "C" , "G"]
  COUNTER = [0 , 0 , 0, 0]
  seq = seq.upper()
  for i in seq:
    if i not in NUC:
      print(i)
      return False
    COUNTER[NUC.index(i)] = COUNTER[NUC.index(i)] + 1
  print(COUNTER)
  return True
```

```python
Count(sequence)
```

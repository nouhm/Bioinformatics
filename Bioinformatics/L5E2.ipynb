```python
# complexity O(n)

def CoinRowDynamicProgramming(C):
  F = [ 0, C[0]]
  n = len(C)
  for i in range (2,n+1):
    F.append(max(C[i-1] + F[i-2] , F[i-1]))
  print(F)
  return F[-1]
```

```python
# CoinRowDynamicProgramming([2,5,10,1,4,3])
# CoinRowDynamicProgramming([5,1,2,10,6,2])
CoinRowDynamicProgramming([7, 9, 10, 9, 3, 5, 2])
```

```python
# complexity O(2^n)
# 2^(n/2)

def CoinRowRecursive(C):

  if len(C) == 1:
    return C[0]
  if len(C) == 0:
    return 0
  
  return max(C[0] + CoinRowRecursive(C[2:]) , CoinRowRecursive(C[1:]))
```

```python
# CoinRowRecursive([5,22,26,10,4,8])
CoinRowRecursive([7, 9, 10, 9, 3, 5, 2])
```

```python
def CoinRowTraceback(C, F):
  coins = []
  flag = False
  for i in reversed(range(1,len(F))):
    if flag:
      C = C[:-1]
      flag = False
      continue
    if F[i] != F[i-1]:
      coins.append(C[-1])
      flag = True
    C = C[:-1]
  # coins.append(C[0])
  print(coins[::-1])
```

```python
# CoinRowTraceback([5,1,2,10,6,2], [0, 5, 5, 7, 15, 15, 17])
CoinRowTraceback([7, 9, 10, 9, 3, 5, 2], [0, 7, 9, 17, 18, 20, 23, 23])
```

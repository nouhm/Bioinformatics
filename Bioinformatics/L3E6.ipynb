```python
def KMER(seq, k):
  kmers = {}
  for i in range(len(seq)-k+1):
    if seq[i:i+k] not in kmers:
      kmers[seq[i:i+k]] = 1
    else:
      kmers[seq[i:i+k]] = kmers[seq[i:i+k]] + 1
  print(kmers)
```

```python
KMER("GATGAT" , 3)
```

```python
def RevKMER(k, kmers):
  seq = ""
  for i in range(len(kmers)-1):
    seq += (kmers[i])[0]
  seq += kmers[-1]
  print(seq)
```

```python
RevKMER(4, ['GGCT', 'GCTT' , 'CTTA' , 'TTAC'])
```

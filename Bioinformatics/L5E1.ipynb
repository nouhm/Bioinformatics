```python
# complexity O(2^n)
# number of operations 2^(n/2)

def FibonacciDivideAndConquer(n):
  if n < 2:
    return n
  return FibonacciDivideAndConquer(n-1)+FibonacciDivideAndConquer(n-2)
```

```python
# complexity O(n)
# number of operations n-2

def FibonacciDynamicProgramming(n):
  array = [0,1]
  for i in range(2,n+1):
    array.append(array[i-1]+array[i-2])
  return array[-1]
```

```python
FibonacciDivideAndConquer(8)
```

```python
FibonacciDynamicProgramming(8)
```

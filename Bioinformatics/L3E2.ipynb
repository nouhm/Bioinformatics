```python
sequence = input("DNA sequence : ")
```

```python
def RNAComplement(seq):
  compSeq = []
  count = 0
  for i in seq:
    if i == "A":
      compSeq.insert(count,"U")
    elif i == "U":
      compSeq.insert(count,"A")
    elif i == "C":
      compSeq.insert(count,"G")
    elif i == "G":
      compSeq.insert(count,"C")
    count = count + 1
  return compSeq
```

```python
def RNAComplement2(seq):
  complement = {"A":"U","U":"A","C":"G","G":"C"}
  seqDash = ""
  for i in seq:
    seqDash += complement[i]
  return seqDash
```

```python
def ReverseComplement(seq):
  return seq[::-1]
```

```python
print(RNAComplement(sequence))
```

```python
print(RNAComplement2(sequence))
```

```python
print(ReverseComplement(sequence))
```

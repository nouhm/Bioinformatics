```python
sequence = input("DNA sequence : ")
```

```python
def Complement(seq):
  compSeq = []
  count = 0
  for i in seq:
    if i == "A":
      compSeq.insert(count,"T")
    elif i == "T":
      compSeq.insert(count,"A")
    elif i == "C":
      compSeq.insert(count,"G")
    elif i == "G":
      compSeq.insert(count,"C")
    count = count + 1
  return compSeq
```

```python
def Complement2(seq):
  complement = {"A":"T","T":"A","C":"G","G":"C"}
  seqDash = ""
  for i in seq:
    seqDash += complement[i]
  return seqDash
```

```python
def ReverseComplement(seq):
  return seq[::-1]
```

```python
print(Complement(sequence))
```

```python
print(Complement2(sequence))
```

```python
print(ReverseComplement(sequence))
```

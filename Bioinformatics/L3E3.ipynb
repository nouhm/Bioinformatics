```python
def DNA2RNA(seq):
  complement = {"A":"U","T":"A","C":"G","G":"C"}
  bases = ""
  for i in seq:
    bases += complement[i]
  print(bases)
```

```python
DNA2RNA("GATCG")
```

```python
def RNA2DNA(seq):
  complement = {"A":"T","U":"A","C":"G","G":"C"}
  bases = ""
  for i in seq:
    bases += complement[i]
  print(bases)
```

```python
RNA2DNA("GAUCG")
```

```python
sequence = input("DNA nucleotide sequence: ") 
```

```python
pattern = input("DNA nucleotide pattern: ") 
```

```python
def PatternCount(sequence, pattern):
  location = []
  counter = 0
  for s in range(len(sequence)-len(pattern)+1):
    if sequence[s : s+len(pattern)] == pattern:
      counter = counter + 1
      location.append(s)
  

  print(counter)
  print(location)
```

```python
PatternCount(sequence, pattern)
```

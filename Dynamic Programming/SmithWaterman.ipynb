```python
def SmithWatermanTraceback(seq1, seq2, matchScore, mismatchScore, gapPenalty, arr, maxI, maxJ):
  output1 = []
  output2 = []

  i = maxI
  j = maxJ

  while arr[i][j] != 0:
    if seq1[j-1] == seq2[i-1] or max( arr[i-1][j-1] , arr[i][j-1] , arr[i-1][j] ) == arr[i-1][j-1] :
      output1.insert(0, seq1[j-1])
      output2.insert(0, seq2[i-1])
      i -= 1
      j -= 1
    elif max( arr[i-1][j-1] , arr[i][j-1] , arr[i-1][j] ) == arr[i][j-1] : 
      output1.insert(0, seq1[j-1])
      output2.insert(0, " ")
      j -= 1
    elif max( arr[i-1][j-1] , arr[i][j-1] , arr[i-1][j] ) == arr[i-1][j] : 
      output2.insert(0, seq2[i-1])      
      output1.insert(0, " ")
      i -= 1

  print(output1)
  print(output2)
```

```python
def SmithWatermanDynProgramming(seq1, seq2, matchScore, mismatchScore, gapPenalty):
  cols = len(seq1)+1
  rows = len(seq2)+1
  arr = [[0 for i in range(cols)] for j in range(rows)]

  maximum = 0
  maxI = 0
  maxJ = 0

  for i in range(1, rows):
    for j in range(1, cols):
      if seq1[j-1] == seq2[i-1]:
        arr[i][j] = max( arr[i-1][j-1] + matchScore, arr[i][j-1] + gapPenalty, arr[i-1][j] + gapPenalty, 0 ) 
        if arr[i][j] > maximum :
          maximum = arr[i][j]
          maxI = i
          maxJ = j
      else: 
        arr[i][j] = max( arr[i-1][j-1] + mismatchScore, arr[i][j-1] + gapPenalty, arr[i-1][j] + gapPenalty, 0 )
        if arr[i][j] > maximum :
          maximum = arr[i][j]
          maxI = i
          maxJ = j

  print(arr) 
  SmithWatermanTraceback(seq1, seq2, matchScore, mismatchScore, gapPenalty, arr, maxI, maxJ)
```

```python
def knapsack(values, weights, maxWeight):

  arr = [[0 for i in range(len(values) + 1)] for j in range(maxWeight + 1) ]

  for i in range(maxWeight + 1):
    for j in range(1, len(values) + 1):

      if weights[j-1] < i:
        arr[i][j] = max( arr[i][j-1] , values[j-1] + arr[ i - weights[j-1] ][j-1] )
      else :
        arr[i][j] = arr[i][j-1]
  print(arr)

  return(arr[-1][-1])
```

```python
value = [ 20, 5, 10, 40, 15, 25 ]
weight = [ 1, 2, 3, 8, 7, 4 ]
W = 10
knapsack(value, weight, W)
```

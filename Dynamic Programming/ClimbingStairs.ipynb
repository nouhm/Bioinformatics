```python
# same as fibonacci

# n = 1 , 1 
# 1

# n = 2 , 2
# 1 1
# 2

# n = 3 , 3
# 1 1 1
# 2 1
# 1 2

# n = 4 , 5
# 1 1 1 1
# 1 2 1
# 1 1 2
# 2 1 1
# 2 2

def climbStairs(n):
  arr = [j for j in range(n+1)]

  for i in range(2, n+1):
    arr[i] = arr[i-1] + arr[i-2]
  print(arr[n])
```

```python
climbStairs(10)
```

```python
def LCSTraceback(x, y , arr):
  i = len(x)
  j = len(y)
  lcs = ""

  while i > 0 and j > 0 :

    if x[i-1] == y[j-1]:
      lcs += x[i-1]
      i = i - 1
      j = j - 1

    elif arr[i][j-1] > arr[i-1][j] :
      j = j - 1
    
    else :
      i = i - 1

  return lcs[::-1]
```

```python
def LCSAllTracebacks(x, y, arr, i, j) :

  if i == 0 or j == 0:
    return ['']

  if x[i-1] == y[j-1]:
    LCS = LCSAllTracebacks(x, y, arr, i-1, j-1) 

    for k in range(len(LCS)) :
      LCS[k] += x[i-1]
    return LCS 

  if arr[i][j-1] > arr[i-1][j] :
    return LCSAllTracebacks(x, y, arr, i, j-1) 

  if arr[i][j-1] < arr[i-1][j] :
    return LCSAllTracebacks(x, y, arr, i-1, j) 

  if arr[i][j-1] == arr[i-1][j] :
    LCS1 = LCSAllTracebacks(x, y, arr, i-1, j) 
    LCS2 = LCSAllTracebacks(x, y, arr, i, j-1) 
    return LCS1 + LCS2
```

```python
def LCSDynamicProgramming(x, y):
  cols = len(y) + 1
  rows = len(x) + 1
  arr = [[0 for i in range(cols)] for j in range(rows)]

  for i in range(1, rows):
    for j in range(1, cols):

      if x[i-1] == y[j-1]:
        arr[i][j] = arr[i-1][j-1] + 1

      else:
        arr[i][j] = max(arr[i-1][j], arr[i][j-1])

  print( "LCS : " + LCSTraceback(x, y , arr) )

  print("All possible LCS : ")
  print( set ( LCSAllTracebacks(x, y, arr, len(x), len(y)) ) )
```

```python
LCSDynamicProgramming("GACTAATA", "ATGACTATAA")
LCSDynamicProgramming("TGACTAC", "ACTGATGC")
LCSDynamicProgramming("ATCCG" , "ATG")
LCSDynamicProgramming("ATCCG" , "ATGCC")
LCSDynamicProgramming("ABCBDAB" , "BDCABA")
```

```python
def LCSRecursive(x, y) :
  if len(x) == 0 or len(y) == 0:
    return 0

  if x[-1] == y[-1]:
    return LCSRecursive(x[:-1], y[:-1]) + 1

  return max( LCSRecursive(x, y[:-1]) , LCSRecursive(x[:-1], y) )
```

```python
LCSRecursive("GACTAATA", "ATGACTATAA")
LCSRecursive("TGACTAC", "ACTGATGC")
LCSRecursive("ATCCG" , "ATG")
LCSRecursive("ATCCG" , "ATGCC")
```

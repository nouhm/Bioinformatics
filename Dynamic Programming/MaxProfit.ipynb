```python
def maxProfit(stocks):
  
  arr = [[0 for i in range(len(stocks))] for j in range(len(stocks))]

  for i in range(len(stocks)):
    for j in range(i+1, len(stocks)):
      arr[i][j] = max(arr[i-1][j-1] , arr[i-1][j] , arr[i][j-1] , stocks[j] - stocks[i])

  print(arr[i][j-1])
```

```python
maxProfit([7,1,5,3,6,4])
maxProfit([7,2,5,3,7,4,1,10])
```

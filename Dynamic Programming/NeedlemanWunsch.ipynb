```python
def NeedlemanWunschTraceback(seq1, seq2, matchScore, mismatchScore, gapPenalty, arr):
  output1 = []
  output2 = []

  i = len(seq1)
  j = len(seq2)

  while i>0 and j>0:
    if seq1[i-1] == seq2[j-1] or max( arr[j-1][i-1] + mismatchScore, arr[j][i-1] + gapPenalty, arr[j-1][i] + gapPenalty ) == arr[j-1][i-1] + mismatchScore :
      output1.insert(0, seq1[i-1])
      output2.insert(0, seq2[j-1])
      i -= 1
      j -= 1
    elif max( arr[j-1][i-1] + mismatchScore, arr[j][i-1] + gapPenalty, arr[j-1][i] + gapPenalty ) == arr[j][i-1] + gapPenalty:
      output1.insert(0, seq1[i-1])
      output2.insert(0, " ")
      i -= 1
    elif max( arr[j-1][i-1] + mismatchScore, arr[j][i-1] + gapPenalty, arr[j-1][i] + gapPenalty ) == arr[j-1][i] + gapPenalty:
      output2.insert(0, seq2[j-1])
      output1.insert(0, " ")
      j -= 1
  
  while i>0 :
    output1.insert(0, seq1[i-1])
    output2.insert(0, " ")
    i -= 1

  while j>0 :
    output2.insert(0, seq2[j-1])
    output1.insert(0, " ")
    j -= 1

  print(output1)
  print(output2)
```

```python
def NeedlemanWunschDynProgramming(seq1, seq2, matchScore, mismatchScore, gapPenalty):
  cols = len(seq1)+1
  rows = len(seq2)+1
  arr = [[0 for i in range(cols)] for j in range(rows)]

  for i in range(1, cols):
    arr[0][i] = arr[0][i-1] + gapPenalty
  
  for i in range(1, rows):
    arr[i][0] = arr[i-1][0] + gapPenalty

  for i in range(1, rows):
    for j in range(1, cols):
      if seq1[j-1] == seq2[i-1]:
        arr[i][j] = max( arr[i-1][j-1] + matchScore, arr[i][j-1] + gapPenalty, arr[i-1][j] + gapPenalty ) 
      else: 
        arr[i][j] = max( arr[i-1][j-1] + mismatchScore, arr[i][j-1] + gapPenalty, arr[i-1][j] + gapPenalty ) 

  
  print(arr)

  NeedlemanWunschTraceback(seq1, seq2, matchScore, mismatchScore, gapPenalty, arr)
```

```python
# same as longest common subsequence but with a string and itself reversed

#     A B B D C A C B
#   0 0 0 0 0 0 0 0 0
# B 0 0 1 1 1 1 1 1 1
# C 0 0 1 1 1 2 2 2 2
# A 0 1 1 1 1 2 3 3 3
# C 0 1 1 1 1 2 3 4 4
# D 0 1 1 1 2 2 3 4 4
# B 0 1 2 2 2 2 3 4 5
# B 0 1 2 3 3 3 3 4 5
# A 0 1 2 3 3 3 4 4 5


# BCACB of length 5
```

```python
def LCSTraceback(seq, arr) :
  rows = len(seq) + 1
  cols = len(seq) + 1

  string1 = seq
  string2 = seq[::-1]

  i = rows - 1
  j = cols - 1

  subsequence = ""

  while i > 0 and j > 0 : 

    #print(cell)
    if string1[j-1] == string2[i-1]:
      subsequence += string1[j-1]
      # print(string1[i-1])
      i = i - 1
      j = j - 1

    elif max( arr[i][j-1] , arr[i-1][j] ) == arr[i][j-1] :
      j = j - 1

    else :
      i = i - 1
      
  return subsequence[::-1]
```

```python
def LongestPalindromeSubsequence(seq):
  rows = len(seq) + 1
  cols = len(seq) + 1
  arr = [[0 for i in range(cols)] for j in range(rows)] 

  string1 = seq
  string2 = seq[::-1]

  for i in range(1, rows):
    for j in range(1, cols):
      if string1[j-1] == string2[i-1] :
        arr[i][j] = arr [i-1][j-1] + 1
      else: 
        arr[i][j] = max(arr[i-1][j], arr[i][j-1])

  print(arr)
  print("Longest Common Subsequence is of length " + str(arr[-1][-1]))

  print("Longest Palindrome Subsequence : " + LCSTraceback(seq, arr))
```

```python
def LongestCommonSubstringTraceback(string1, string2, arr, row):

  i = len(string1)
  j = row
  lcs = ""

  while i > 0 and j > 0:

    if string1[i-1] == string2[j-1]:
      lcs += string1[i-1]
      i = i - 1
      j = j - 1

    elif arr[i][j-1] > arr[i-1][j] :
      j = j - 1
    
    else :
      i = i - 1

  return lcs[::-1]
```

```python
def LongestCommonSubstring(string1, string2):
  
  cols = len(string1) + 1
  rows = len(string2) + 1
  arr = [[0 for i in range(cols)] for j in range(rows)] 
  
  max = 0
  row = 0

  for i in range(len(string1)):
    for j in range(len(string2)):

      if string1[i] == string2[j]:
        arr[i+1][j+1] = arr[i][j] + 1

        if arr[i+1][j+1] > max : 
          max = arr[i+1][j+1]
          row = j+1

      else :
        arr[i+1][j+1] = 0

  print("Longest common substring length : " + str( max ))
  print("Longest common substring : " + LongestCommonSubstringTraceback(string1, string2, arr, row))
```

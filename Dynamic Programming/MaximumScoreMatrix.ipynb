```python
def MaxScoreMatrixTraceback(seq1, seq2, matchScore, mismatchScore, gapPenalty, arr):
  cols = len(seq1)
  rows = len(seq2)

  text1 = [" " for i in range(arr[-1][-1]+1)]
  text2 = [" " for i in range(arr[-1][-1]+1)]

  i = cols - 1
  j = rows - 1

  for cell in reversed(range(0,arr[-1][-1]+1)):

    if seq1[i] == seq2[j] or max( arr[i-1][j-1] + mismatchScore, arr[i][j-1] + gapPenalty, arr[i-1][j] + gapPenalty ) == arr[i-1][j-1] + mismatchScore:
      text1[cell] = seq1[i]
      text2[cell] = seq2[j]
      i = i-1
      j = j-1
    elif max( arr[i-1][j-1] + mismatchScore, arr[i][j-1] + gapPenalty, arr[i-1][j] + gapPenalty ) == arr[i][j-1] + gapPenalty:
      text2[cell] = seq2[i]
      j = j-1
    elif max( arr[i-1][j-1] + mismatchScore, arr[i][j-1] + gapPenalty, arr[i-1][j] + gapPenalty ) == arr[i-1][j] + gapPenalty:
      text1[cell] = seq1[i]
      i = i-1
      
  print(text1)
  print(text2)
```

```python
def MaxScoreMatrixDynProgramming(seq1, seq2, matchScore, mismatchScore, gapPenalty):
  cols = len(seq1)
  rows = len(seq2)
  arr = [[0 for i in range(cols)] for j in range(rows)]

  for i in range(0, cols):
    if seq1[i] == seq2[0]:
      arr[0][i] = 1
    else :
      arr[0][i] = 0
  
  for i in range(0, rows):
    if seq1[0] == seq2[i]:
      arr[i][0] = 1
    else :
      arr[i][0] = 0

  for i in range(1, rows):
    for j in range(1, cols):
      if seq1[j] == seq2[i]:
        arr[i][j] = max( arr[i-1][j-1], arr[i][j-1], arr[i-1][j]) + matchScore
      else: 
        arr[i][j] = max( arr[i-1][j-1] + mismatchScore, arr[i][j-1] + gapPenalty, arr[i-1][j] + gapPenalty ) 

  
  print(arr)

  MaxScoreMatrixTraceback(seq1, seq2, matchScore, mismatchScore, gapPenalty, arr)
```

```python
MaxScoreMatrixDynProgramming("ATTG", "AGG", 1, 0, 0)
MaxScoreMatrixDynProgramming("CATGTT", "CACTGT", 1, 0, 0)
```

```python
def LinearGap(seq1, seq2, matchScore, mismatchScore, gapPenalty):
  
  score = 0
  
  for i in range(len(seq1)):
    if seq1[i] == seq2[i]:
      score += matchScore
    elif seq1[i] == " " or seq2[i] == " " :
      score += gapPenalty
    else:
      score += mismatchScore
  
  return score
```

```python
def AffineGap(seq1, seq2, matchScore, mismatchScore, gapOpening, gapExtension):
  
  score = 0
  i = 0
  
  while i < len(seq1):
    count = 1
    if seq1[i] == seq2[i]:
      score += matchScore
    # i = 3
    elif seq1[i] == " " :
      score += gapOpening
      while ( (i < len(seq1) - 1) and seq1[i+1] == " "):
        i += 1
        count += 1
      score += gapExtension * count
    elif seq2[i] == " " :
      score += gapOpening
      while ( (i < len(seq2) - 1) and seq2[i+1] == " "):
        i += 1
        count += 1
      score += gapExtension * count
    else:
      score += mismatchScore
    
    i += 1
  
  return score
```

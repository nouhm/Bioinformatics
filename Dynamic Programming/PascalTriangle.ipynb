```python
def PascalTriangle(n):

  arr = [[1 for i in range(n)] for j in range(n)]

  for i in range(n):
    for j in range(1, i):
      arr[i][j] = arr[i-1][j-1] + arr[i-1][j]

  for i in range(n):
    for j in range(i + 1):
      print(arr[i][j], end = " ")
    print("\n")
```

```python
PascalTriangle(6)
```

```python
def LCSTraceback(seq, arr, row) :
  rows = len(seq) + 1
  cols = len(seq) + 1

  string1 = seq
  string2 = seq

  i = row
  j = cols - 1

  subsequence = ""

  while i > 0 and j > 0 : 

    if string1[j-1] == string2[i-1]:
      subsequence += string1[j-1]
      i = i - 1
      j = j - 1

    elif max( arr[i][j-1] , arr[i-1][j] ) == arr[i][j-1] :
      j = j - 1

    else :
      i = i - 1

  return subsequence[::-1]
```

```python
def LongestRepeatedSubsequence(seq):
  rows = len(seq) + 1
  cols = len(seq) + 1
  arr = [[0 for i in range(cols)] for j in range(rows)] 

  string1 = seq
  string2 = seq

  row = 0
  maximum = 0

  for i in range(1, rows):
    for j in range(1, cols):
      if i == j :
        arr[i][j] = 0

      elif string1[j-1] == string2[i-1] :
        arr[i][j] = arr[i-1][j-1] + 1
        if arr[i][j] > maximum :
          maximum = arr[i][j]
          row = j

      else: 
        arr[i][j] = max( arr[i-1][j] , arr[i][j-1] )

  # print(arr)
  print("Longest Common Subsequence is of length " + str( maximum ))

  print("Longest Repeated Subsequence : " + LCSTraceback(seq, arr, row))
```
